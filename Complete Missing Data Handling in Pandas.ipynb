```python
import pandas as pd
import numpy as np

# ==============================
# 1️⃣ Create Dataset with Missing Values
# ==============================

data = {
    'First Score': [100, 90, np.nan, 95],
    'Second Score': [30, 45, 56, np.nan],
    'Third Score': [np.nan, 40, 80, 98]
}

df = pd.DataFrame(data)

print("===== Original Data =====")
print(df)

# ==============================
# 2️⃣ Detect Missing Values
# ==============================

print("\n===== Missing Values (isnull) =====")
print(df.isnull())

print("\n===== Missing Values (isna) =====")
print(df.isna())

print("\n===== Non-Missing Values (notnull) =====")
print(df.notnull())

print("\n===== Data Info =====")
df.info()

# ==============================
# 3️⃣ Fill Missing Values
# ==============================

print("\n===== Fill with 0 =====")
print(df.fillna(0))

print("\n===== Forward Fill =====")
print(df.fillna(method='ffill'))

print("\n===== Backward Fill =====")
print(df.fillna(method='bfill'))

print("\n===== Replace NaN with -99 =====")
print(df.replace(to_replace=np.nan, value=-99))

# ==============================
# 4️⃣ Interpolation
# ==============================

print("\n===== Interpolated Data (Linear) =====")
print(df.interpolate(method='linear'))

# ==============================
# 5️⃣ Drop Missing Values
# ==============================

print("\n===== Drop Rows with Any NaN =====")
print(df.dropna())

print("\n===== Drop Rows with All NaN =====")
print(df.dropna(how='all'))

print("\n===== Drop Columns with Any NaN =====")
print(df.dropna(axis=1))

# ==============================
# 6️⃣ Remove Duplicates Example
# ==============================

df_dup = df.copy()
df_dup.loc[4] = df_dup.loc[0]  # Add duplicate row

print("\n===== Data with Duplicate Row =====")
print(df_dup)

print("\n===== After Removing Duplicates =====")
print(df_dup.drop_duplicates())

# ==============================
# 7️⃣ Unique Values Example
# ==============================

print("\n===== Unique Values in First Score Column =====")
print(df['First Score'].unique())
```
